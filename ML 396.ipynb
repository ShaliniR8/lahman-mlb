```python
def get_win_perc (row):
    year = row['yearID'] - 1
    team = row['teamID']
    filtered_df = fullteamsdf[(fullteamsdf['yearID'] == year) & (fullteamsdf['teamID'] == team)]
    if filtered_df.empty:
        return 'NaN'
    w = filtered_df.iloc[0, 8]
    l = filtered_df.iloc[0, 9]
    return w / (w + l)
```

```python
def win_perc (team, year):
    filtered_df = fullteamsdf[(fullteamsdf['yearID'] == year) & (fullteamsdf['teamID'] == team)]
    if filtered_df.empty:
        return 'NaN'
    w = filtered_df.iloc[0, 8]
    l = filtered_df.iloc[0, 9]
    return w / (w + l)
def get_rolling_5_win_perc (row):
    year = row['yearID']
    team = row['teamID']
    year1 = win_perc(team, year - 1)
    year2 = win_perc(team, year - 2)
    year3 = win_perc(team, year - 3)
    year4 = win_perc(team, year - 4)
    year5 = win_perc(team, year - 5)
    if (year1 == 'NaN' or year2 == 'NaN' or year3 == 'NaN' or year3 == 'NaN' or year4 == 'NaN' or year5 == 'NaN'):
        return 'NaN'
    else: 
        return (year1 + year2 + year3 + year4 + year5)/5
```

```python
features_df['5_year_win_perc'] = features_df.apply (lambda row: get_rolling_5_win_perc(row), axis=1)
```
