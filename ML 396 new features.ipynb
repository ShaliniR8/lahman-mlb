```python
def win_perc (team, year):
    filtered_df = fullteamsdf[(fullteamsdf['yearID'] == year) & (fullteamsdf['teamID'] == team)]
    if filtered_df.empty:
        return 'NaN'
    w = filtered_df.iloc[0, 8]
    l = filtered_df.iloc[0, 9]
    return w / (w + l)
def get_rolling_5_win_perc (row):
    year = row['yearID']
    team = row['teamID']
    year1 = win_perc(team, year - 1)
    year2 = win_perc(team, year - 2)
    year3 = win_perc(team, year - 3)
    year4 = win_perc(team, year - 4)
    year5 = win_perc(team, year - 5)
    if (year1 == 'NaN' or year2 == 'NaN' or year3 == 'NaN' or year3 == 'NaN' or year4 == 'NaN' or year5 == 'NaN'):
        return 'NaN'
    else: 
        return (year1 + year2 + year3 + year4 + year5)/5
```

```python
features_df['5_year_win_perc'] = features_df.apply (lambda row: get_rolling_5_win_perc(row), axis=1)
```

```python
def get_rolling_3_win_perc (row):
    year = row['yearID']
    team = row['teamID']
    year1 = win_perc(team, year - 1)
    year2 = win_perc(team, year - 2)
    year3 = win_perc(team, year - 3)
    if (year1 == 'NaN' or year2 == 'NaN' or year3 == 'NaN' or year3 == 'NaN'):
        return 'NaN'
    else: 
        return (year1 + year2 + year3)/3
```

```python
def ops_calc(row):
    #calc done on rows of batting table
    if row['AB'] == 0:
        return 'NaN'
    if pd.isnull(row["HBP"]):
        HBP = 0
    else:
        HBP = row["HBP"]
    if pd.isnull(row["SF"]):
        SF = 0
    else:
        SF = row["SF"]
    singles = row["H"] - row["2B"] - row["3B"] - row["HR"]
    obp = (row["H"] + row["BB"] + HBP) / (row["AB"] + row["BB"] + SF + HBP)
    slug = (singles + (2 * row["2B"]) + (3 * row["3B"]) + (4 * row["HR"])) / row["AB"]
    return slug + obp 
```

```python
def get_ops(row):
    year = row['yearID'] - 1
    team = row['teamID']
    cols = ["H", "BB", "HBP", "AB", "SF", "2B", "3B", "HR"]
    filtered_df = batting[(batting['yearID'] == year) & (batting['teamID'] == team)][cols]
    filtered_df = filtered_df.dropna()
    if filtered_df.empty:
        return 'NaN'
    filtered_df["ops"] = filtered_df.apply(lambda row: ops_calc(row), axis=1)
    filtered_df = filtered_df.dropna()
    if filtered_df[["ops"]].empty:
        return 'NaN'
    else: 
        return filtered_df["ops"].median()

    
    
```

```python
batting["ops"] = batting.apply(lambda row: ops_calc(row), axis=1)
```

```python
def get_median_ops(row):
    year = row['yearID'] - 1
    team = row['teamID']
    filtered_df = batting[(batting['yearID'] == year) & (batting['teamID'] == team)]
    filtered_df = filtered_df[filtered_df['ops'] != 'NaN']
    if filtered_df.empty:
        return 'NaN'
    return filtered_df['ops'].median()
```
